```python
import pandas as pd
```

```python
df=pd.read_csv('testdata_with_predictions6.csv')
df.rename(columns={'文件夹': 'folder', '文件名': 'file'}, inplace=True)
df = df.sort_values(by=['folder', 'file'])
df
```

```python
grouped = df.groupby(['folder', 'file'])['nofreeze_label'].value_counts(normalize=True)

# 将结果恢复为 DataFrame
result = grouped.rename('nofree_proportion').reset_index()
```

```python
result.rename(columns={'nofreeze_label': 'label'}, inplace=True)
result
```

```python
grouped = df.groupby(['folder', 'file'])['half_freeze_label'].value_counts(normalize=True)

# 将结果恢复为 DataFrame
result2 = grouped.rename('half_freeze_proportion').reset_index()
result2.rename(columns={'half_freeze_label': 'label'}, inplace=True)
result=pd.merge(result,result2,on=['folder'	,'file','label'], how='outer')
```

```python
grouped = df.groupby(['folder', 'file'])['freeze_label'].value_counts(normalize=True)

# 将结果恢复为 DataFrame
result2 = grouped.rename('freeze_proportion').reset_index()
result2.rename(columns={'freeze_label': 'label'}, inplace=True)
result=pd.merge(result,result2,on=['folder'	,'file','label'], how='outer')
```

```python
result.fillna(0,inplace=True)
result = result[result['folder'] != 1102]

result['file'] = result['file'].str.replace('.txt', '', regex=False)
result['file'] = result['file'].astype(int)
```

```python

result.sort_values(by=['folder', 'file'],inplace=True)
result
```

```python
all_labels = [0, 1, 2, 3, 4, 5]

# 按 folder 和 file 分组
groups = result.groupby(['folder', 'file'])

# 创建一个空的结果 DataFrame
res = pd.DataFrame()

for (folder, file), group in groups:
    # 找到缺失的 label
    missing_labels = set(all_labels) - set(group['label'])
    
    # 创建缺失行
    if missing_labels:
        missing_data = pd.DataFrame({
            'folder': [folder] * len(missing_labels),
            'file': [file] * len(missing_labels),
            'label': list(missing_labels),
        })
        for col in df.columns:
            if col not in ['folder', 'file', 'label']:
                missing_data[col] = 0
        # 合并缺失行和原始组
        group = pd.concat([group, missing_data], ignore_index=True)
    
    # 添加到结果 DataFrame
    res = pd.concat([res, group], ignore_index=True)

# 按 folder, file 和 label 排序
res = res.sort_values(by=['folder', 'file', 'label']).reset_index(drop=True)

res
```

```python
yvalue=pd.read_csv('yvalue.csv')
yvalue
```

```python
yvalue['Text'] = yvalue['Text'].str.replace(r'^(0%|20%|40%|60%|80%|100%)\n', '', regex=True)
```

```python

# 使用正则表达式提取所有 label 和 ratio
yvalue_expanded = (
    yvalue['Text']
    .str.extractall(r'([^\n]+)\n(\d+)%')  # 提取 label 和 ratio
    .rename(columns={0: 'label', 1: 'ratio'})  # 重命名列
)

# 将 ratio 转换为数值类型
yvalue_expanded['ratio'] = yvalue_expanded['ratio'].astype(int)

# 保留原始行信息，展开后每行对应相应的 label 和 ratio
yvalue_expanded.reset_index(level=1, drop=True, inplace=True)  # 删除多重索引的第二级
yvalue_expanded = yvalue.drop(columns=['Text']).join(yvalue_expanded)  # 将原始列与新数据合并

# 查看结果
```
