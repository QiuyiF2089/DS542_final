```python
from paddleocr import PaddleOCR
import os
import pandas as pd

def extract_text_with_paddleocr(base_folder):
    # 初始化 PaddleOCR
    ocr = PaddleOCR(use_angle_cls=True, lang="ch")  # 支持中文识别
    results = []

    # 遍历文件夹和图片
    for folder_name in os.listdir(base_folder):
        folder_path = os.path.join(base_folder, folder_name)
        if os.path.isdir(folder_path):
            for image_name in os.listdir(folder_path):
                image_path = os.path.join(folder_path, image_name)
                try:
                    # 使用 PaddleOCR 进行文字识别
                    result = ocr.ocr(image_path, cls=True)
                    # 提取文字部分
                    text = "\n".join([line[1][0] for line in result[0]])
                    results.append({"Folder": folder_name, "Image": image_name, "Text": text})
                except Exception as e:
                    print(f"Error processing {image_path}: {e}")

    # 保存为 CSV 文件
    df = pd.DataFrame(results)
    output_path = os.path.join(base_folder, "yvalue.csv")
    df.to_csv(output_path, index=False, encoding="utf-8-sig")
    print(f"数据已提取并保存到 {output_path}")

# 替换为你的主文件夹路径
base_folder = "image"
extract_text_with_paddleocr(base_folder)
```
